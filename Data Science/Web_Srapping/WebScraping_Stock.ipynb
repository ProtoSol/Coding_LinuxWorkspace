```python
# Import the Libraries
import yfinance as yf
import pandas as pd

# Define the Ticker Symbol
tickerSymbol = 'AAPL'
```

```python
# get the data of the Specific ticker
tickerData = yf.Ticker(tickerSymbol)

# Print out if running
# tickerData.info
```

```python
# get the Historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2010-1-1', end='2020-1-25')

# Last Closing Price
tickerDf['Close'].iloc[-1], tickerDf.shape
```

## Fetch the Stock Price data using Realtime Time and Date

```python
import pandas as pd
import yfinance as yf
import datetime as dict
from datetime import date, timedelta

today = date.today()
d1 = today.strftime("%Y-%m-%d")
d1
d2 = (today - timedelta(days=365)).strftime("%Y-%m-%d")
d2

start_date = d2
end_date = d1

# define the Ticker Symbol
tickerSymbol = 'GOOGL'
# get the data of the Specific ticker
tickerData = yf.Ticker(tickerSymbol)
# get the Historical prices for this ticker
tickerDf = tickerData.history(period='1d', start=start_date, end=end_date)
# Last Closing Price
tickerDf.head()
```
